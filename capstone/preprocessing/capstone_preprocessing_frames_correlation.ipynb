```python
import os
import matplotlib.pyplot as plt
import numpy as np
import cv2
import scipy
from scipy.stats import pearsonr
```

```python
gray_scale=[]
path=r'E:\capstone\code\WLASL-master\WLASL-master\start_kit\preprocessed_vid_by_word_48_64 16.09.58\a\train'
frames=[]
similarity=[]
for vid in os.listdir(path):
    print("vid id",vid)
    vpath=os.path.join(path,vid)
    x_ind=np.load(vpath)
    prev_frame=np.zeros(2)
    for curr_frame in x_ind:
        curr_frame=cv2.cvtColor(curr_frame, cv2.COLOR_BGR2GRAY)
        if prev_frame.shape[0]==2:
            
            prev_frame=curr_frame
            continue
        A=prev_frame.reshape(128*96*1)
        B=curr_frame.reshape(128*96*1)
        
        #print(A)
        #print(B)
        
        frames.append(curr_frame)
        similarity.append(pearsonr(A,B)[0])
        
        prev_frame=curr_frame
        
        
               
```

```python
plt.figure(figsize=(25,25))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(frames[i])
    plt.gca().set_title(similarity[i])
```

```python
plt.figure(figsize=(25,25))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(frames[i+100])
    plt.gca().set_title(similarity[i+100])
```

```python
gray_scale=[]
path=r'E:\capstone\code\WLASL-master\WLASL-master\start_kit\preprocessed_vid_by_word_48_64 16.09.58\all\train'
frames=[]
similarity=[]
for vid in os.listdir(path):
    print("vid id",vid)
    vpath=os.path.join(path,vid)
    x_ind=np.load(vpath)
    prev_frame=np.zeros(2)
    for curr_frame in x_ind:
        curr_frame=cv2.cvtColor(curr_frame, cv2.COLOR_BGR2GRAY)
        if prev_frame.shape[0]==2:
            print("here")
            frames.append(curr_frame)
            similarity.append(0)
            prev_frame=curr_frame
            continue
        A=prev_frame.reshape(128*96*1)
        B=curr_frame.reshape(128*96*1)
        
        #print(A)
        #print(B)
        sim=scipy.stats.pearsonr(A,B)[0]
        if(sim<.95):
            frames.append(curr_frame)
            similarity.append(sim)
            prev_frame=curr_frame
        
        
               
```

```python
plt.figure(figsize=(35,35))
for i in range(49):
    plt.subplot(7,7,i+1)
    plt.imshow(frames[100+i])
    plt.gca().set_title(similarity[100+i])
```
